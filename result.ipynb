```python
import os
import numpy as np
def load_flist(flist):
    if isinstance(flist, list):
        return flist
    print("loading ： %s"%(flist))
    # flist: image file path, image directory path, text file flist path
    if isinstance(flist, str):
        if os.path.isdir(flist):
            flist = list(glob.glob(flist + '/*.jpg')) + list(glob.glob(flist + '/*.png'))
            flist.sort()
            print("len is : %d "%(len(flist)))
            return flist

        if os.path.isfile(flist):
            try:
                return np.genfromtxt(flist, dtype=np.str, encoding='utf-8')
            except Exception as e:
                print(e)
                return [flist]

    return []

test_data = load_flist("datasets/ffhq/test/images.flist")
```

```python
root_path = "/data/chengbin/ffhq-dataset/ffhq-lafin/images"
from tqdm import tqdm
import matplotlib.pyplot as plt
import cv2

for data in tqdm(test_data):
    imgs = cv2.imread(os.path.join(root_path,data))
    plt.imshow(imgs)
    resize_imgs =cv2.resize(imgs,(256,256),interpolation=cv2.INTER_CUBIC)
    cv2.imwrite(os.path.join("checkpoints/test_data_256",data),resize_imgs)
#     os.system("cp %s checkpoints/test_data/"%os.path.join(root_path,data))
```
